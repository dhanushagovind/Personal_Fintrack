```python

import matplotlib.pyplot as plt
import mysql.connector
import gradio as gr
import pandas as pd


# ------------------------------------------------
# DATABASE CONNECTION
# ------------------------------------------------
# Connect Python to MySQL database

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Drace@25",
    database="fintrack"
)

cursor = conn.cursor()


# ------------------------------------------------
# FUNCTION: ADD NEW TRANSACTION
# ------------------------------------------------
# Inserts income or expense into transactions table
# Overspending trigger in MySQL will run automatically

def add_transaction(category_id, t_type, amount, description, date):

    try:
        query = """
        INSERT INTO transactions
        (category_id, type, amount, description, date)
        VALUES (%s, %s, %s, %s, %s)
        """

        cursor.execute(query, (category_id, t_type, amount, description, date))
        conn.commit()

        return "Transaction added successfully"

    except Exception as e:
        return f"Error: {e}"


# ------------------------------------------------
# FUNCTION: VIEW ALL TRANSACTIONS
# ------------------------------------------------
# Retrieves all records from transactions table

def view_transactions():

    try:
        cursor.execute("""
        SELECT transaction_id, category_id, type, amount, description, date
        FROM transactions
        """)

        data = cursor.fetchall()

        df = pd.DataFrame(data, columns=[
            "Transaction_ID",
            "Category_ID",
            "Type",
            "Amount",
            "Description",
            "Date"
        ])

        return df

    except Exception as e:
        return pd.DataFrame({"Error": [str(e)]})


# ------------------------------------------------
# FUNCTION: MONTHLY EXPENSE ANALYTICS
# ------------------------------------------------
# Uses SQL VIEW: monthly_expense_summary

def show_monthly_expense():

    try:
        cursor.execute("""
        SELECT month, category_id, total_expense
        FROM monthly_expense_summary
        """)

        data = cursor.fetchall()

        df = pd.DataFrame(data, columns=[
            "Month",
            "Category_ID",
            "Total_Expense"
        ])

        return df

    except Exception as e:
        return pd.DataFrame({"Error": [str(e)]})


# ------------------------------------------------
# FUNCTION: CATEGORY SPENDING ANALYSIS
# ------------------------------------------------
# Uses SQL VIEW: category_spending

def category_spending():

    try:
        cursor.execute("""
        SELECT category_name, total_spent
        FROM category_spending
        """)

        data = cursor.fetchall()

        df = pd.DataFrame(data, columns=[
            "Category",
            "Total_Spent"
        ])

        return df

    except Exception as e:
        return pd.DataFrame({"Error": [str(e)]})


# ------------------------------------------------
# FUNCTION: SHOW OVERSPENDING ALERTS
# ------------------------------------------------
# Trigger inserts records here when spending exceeds budget

def overspending_alerts():

    try:
        cursor.execute("""
        SELECT alert_id, category_id, total_spent, budget, alert_date
        FROM overspending_alerts
        """)

        data = cursor.fetchall()

        df = pd.DataFrame(data, columns=[
            "Alert_ID",
            "Category_ID",
            "Total_Spent",
            "Budget",
            "Alert_Date"
        ])

        return df

    except Exception as e:
        return pd.DataFrame({"Error": [str(e)]})


# ------------------------------------------------
# FUNCTION: DETECT UNUSUAL TRANSACTIONS
# ------------------------------------------------
# Uses SQL VIEW: unusual_transactions

def unusual_transactions():

    try:
        cursor.execute("""
        SELECT transaction_id, category_id, type, amount, description, date
        FROM unusual_transactions
        """)

        data = cursor.fetchall()

        df = pd.DataFrame(data, columns=[
            "Transaction_ID",
            "Category_ID",
            "Type",
            "Amount",
            "Description",
            "Date"
        ])

        return df

    except Exception as e:
        return pd.DataFrame({"Error": [str(e)]})


# ------------------------------------------------
# GRADIO USER INTERFACE
# ------------------------------------------------

with gr.Blocks() as demo:

    gr.Markdown("FinTrack Personal Finance Dashboard")


    # ------------------------------------------------
    # TAB 1: ADD TRANSACTION
    # ------------------------------------------------

    with gr.Tab("Add Transaction"):

        category_id = gr.Number(label="Category ID")

        t_type = gr.Dropdown(
            choices=["expense", "income"],
            label="Transaction Type"
        )

        amount = gr.Number(label="Amount")

        description = gr.Textbox(label="Description")

        date = gr.Textbox(label="Date (YYYY-MM-DD)")

        add_btn = gr.Button("Add Transaction")

        add_output = gr.Textbox(label="Status")

        add_btn.click(
            fn=add_transaction,
            inputs=[category_id, t_type, amount, description, date],
            outputs=add_output
        )


    # ------------------------------------------------
    # TAB 2: VIEW ALL TRANSACTIONS
    # ------------------------------------------------

    with gr.Tab("View Transactions"):

        view_btn = gr.Button("Load Transactions")

        transaction_table = gr.Dataframe()

        view_btn.click(
            fn=view_transactions,
            outputs=transaction_table
        )


    # ------------------------------------------------
    # TAB 3: MONTHLY EXPENSE ANALYTICS
    # ------------------------------------------------

    with gr.Tab("Monthly Expense Summary"):

        monthly_btn = gr.Button("Generate Monthly Report")

        monthly_table = gr.Dataframe()

        monthly_btn.click(
            fn=show_monthly_expense,
            outputs=monthly_table
        )


    # ------------------------------------------------
    # TAB 4: CATEGORY SPENDING ANALYSIS
    # ------------------------------------------------

    with gr.Tab("Category Spending"):

        category_btn = gr.Button("Analyze Category Spending")

        category_table = gr.Dataframe()

        category_btn.click(
            fn=category_spending,
            outputs=category_table
        )


    # ------------------------------------------------
    # TAB 5: OVERSPENDING ALERTS
    # ------------------------------------------------

    with gr.Tab("Overspending Alerts"):

        alert_btn = gr.Button("Check Alerts")

        alert_table = gr.Dataframe()

        alert_btn.click(
            fn=overspending_alerts,
            outputs=alert_table
        )


    # ------------------------------------------------
    # TAB 6: UNUSUAL SPENDING DETECTION
    # ------------------------------------------------

    with gr.Tab("Unusual Transactions"):

        unusual_btn = gr.Button("Detect Unusual Spending")

        unusual_table = gr.Dataframe()

        unusual_btn.click(
            fn=unusual_transactions,
            outputs=unusual_table)


# ------------------------------------------------
# START APPLICATION
# ------------------------------------------------

demo.launch()
```

```python
import mysql.connector
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt


# ------------------------------------------------
# DATABASE CONNECTION
# Connect Python to MySQL database
# ------------------------------------------------

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Drace@25",
    database="fintrack"
)

cursor = conn.cursor()


# ------------------------------------------------
# FUNCTION TO ADD A NEW TRANSACTION
# Inserts income or expense into transactions table
# ------------------------------------------------

def add_transaction(category_id, t_type, amount, description, date):

    try:
        query = """
        INSERT INTO transactions
        (category_id, type, amount, description, date)
        VALUES (%s,%s,%s,%s,%s)
        """

        cursor.execute(query,(category_id,t_type,amount,description,date))
        conn.commit()

        return "Transaction added successfully"

    except Exception as e:
        return str(e)


# ------------------------------------------------
# FUNCTION TO VIEW ALL TRANSACTIONS
# Retrieves records from transactions table
# ------------------------------------------------

def view_transactions():

    cursor.execute("""
    SELECT transaction_id, category_id, type, amount, description, date
    FROM transactions
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=[
        "Transaction_ID",
        "Category_ID",
        "Type",
        "Amount",
        "Description",
        "Date"
    ])

    return df


# ------------------------------------------------
# FUNCTION TO SHOW MONTHLY EXPENSE SUMMARY
# Uses SQL view: monthly_expense_summary
# ------------------------------------------------

def show_monthly_expense():

    cursor.execute("""
    SELECT month, category_id, total_expense
    FROM monthly_expense_summary
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=[
        "Month",
        "Category_ID",
        "Total_Expense"
    ])

    return df


# ------------------------------------------------
# FUNCTION TO SHOW CATEGORY SPENDING
# Uses SQL view: category_spending
# ------------------------------------------------

def category_spending():

    cursor.execute("""
    SELECT category_name, total_spent
    FROM category_spending
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=[
        "Category",
        "Total_Spent"
    ])

    return df


# ------------------------------------------------
# FUNCTION TO DISPLAY OVERSPENDING ALERTS
# Data generated by MySQL trigger
# ------------------------------------------------

def overspending_alerts():

    cursor.execute("""
    SELECT alert_id, category_id, total_spent, budget, alert_date
    FROM overspending_alerts
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=[
        "Alert_ID",
        "Category_ID",
        "Total_Spent",
        "Budget",
        "Alert_Date"
    ])

    return df


# ------------------------------------------------
# FUNCTION TO DETECT UNUSUAL TRANSACTIONS
# Uses SQL view: unusual_transactions
# ------------------------------------------------

def unusual_transactions():

    cursor.execute("""
    SELECT transaction_id, category_id, type, amount, description, date
    FROM unusual_transactions
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=[
        "Transaction_ID",
        "Category_ID",
        "Type",
        "Amount",
        "Description",
        "Date"
    ])

    return df

# Function to generate budget recommendation

def budget_recommendation():

    cursor.execute("""
    SELECT category_name, avg_spending, recommended_budget
    FROM budget_recommendation
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data, columns=[
        "Category",
        "Average_Spending",
        "Recommended_Budget"
    ])

    return df

# ------------------------------------------------
# FUNCTION TO CREATE CATEGORY EXPENSE PIE CHART
# Data comes from category_spending view
# ------------------------------------------------

def category_pie_chart():

    cursor.execute("""
    SELECT category_name, total_spent
    FROM category_spending
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=["Category","Total"])

    plt.figure()

    plt.pie(
        df["Total"],
        labels=df["Category"],
        autopct="%1.1f%%"
    )

    plt.title("Expense Distribution by Category")

    return plt.gcf()


# ------------------------------------------------
# FUNCTION TO CREATE MONTHLY EXPENSE TREND GRAPH
# Data comes from monthly_expense_summary view
# ------------------------------------------------

def monthly_trend_chart():

    cursor.execute("""
    SELECT month, SUM(total_expense)
    FROM monthly_expense_summary
    GROUP BY month
    """)

    data = cursor.fetchall()

    df = pd.DataFrame(data,columns=["Month","Total_Expense"])

    plt.figure()

    plt.plot(
        df["Month"],
        df["Total_Expense"],
        marker="o"
    )

    plt.xlabel("Month")
    plt.ylabel("Total Expense")
    plt.title("Monthly Expense Trend")

    return plt.gcf()


# ------------------------------------------------
# GRADIO DASHBOARD INTERFACE
# ------------------------------------------------

with gr.Blocks() as demo:

    gr.Markdown("FinTrack Personal Finance Dashboard")


    # TAB FOR ADDING TRANSACTION

    with gr.Tab("Add Transaction"):

        category_id = gr.Number(label="Category ID")
        t_type = gr.Dropdown(["expense","income"],label="Transaction Type")
        amount = gr.Number(label="Amount")
        description = gr.Textbox(label="Description")
        date = gr.Textbox(label="Date (YYYY-MM-DD)")

        add_btn = gr.Button("Add Transaction")
        add_result = gr.Textbox(label="Status")

        add_btn.click(
            fn=add_transaction,
            inputs=[category_id,t_type,amount,description,date],
            outputs=add_result
        )


    # TAB TO VIEW TRANSACTIONS

    with gr.Tab("View Transactions"):

        view_btn = gr.Button("Show Transactions")
        transaction_table = gr.Dataframe()

        view_btn.click(
            fn=view_transactions,
            outputs=transaction_table
        )


    # TAB FOR MONTHLY EXPENSE SUMMARY

    with gr.Tab("Monthly Expense Summary"):

        monthly_btn = gr.Button("Generate Monthly Summary")
        monthly_table = gr.Dataframe()

        monthly_btn.click(
            fn=show_monthly_expense,
            outputs=monthly_table
        )


    # TAB FOR CATEGORY SPENDING ANALYSIS

    with gr.Tab("Category Spending"):

        category_btn = gr.Button("Analyze Category Spending")
        category_table = gr.Dataframe()

        category_btn.click(
            fn=category_spending,
            outputs=category_table
        )


    # TAB FOR OVERSPENDING ALERTS

    with gr.Tab("Overspending Alerts"):

        alert_btn = gr.Button("Check Alerts")
        alert_table = gr.Dataframe()

        alert_btn.click(
            fn=overspending_alerts,
            outputs=alert_table
        )


    # TAB FOR UNUSUAL TRANSACTIONS

    with gr.Tab("Unusual Transactions"):

        unusual_btn = gr.Button("Detect Unusual Spending")
        unusual_table = gr.Dataframe()

        unusual_btn.click(
            fn=unusual_transactions,
            outputs=unusual_table
        )


    # TAB FOR FINANCIAL CHARTS

    with gr.Tab("Financial Charts"):

        pie_btn = gr.Button("Generate Category Expense Pie Chart")
        pie_plot = gr.Plot()

        pie_btn.click(
            fn=category_pie_chart,
            outputs=pie_plot
        )

        trend_btn = gr.Button("Generate Monthly Expense Trend")
        trend_plot = gr.Plot()

        trend_btn.click(
            fn=monthly_trend_chart,
            outputs=trend_plot
        )

# Tab for automatic budget recommendation

    with gr.Tab("Budget Recommendation"):

        recommend_btn = gr.Button("Generate Recommended Budget")

        recommend_table = gr.Dataframe()

        recommend_btn.click(
        fn=budget_recommendation,
        outputs=recommend_table
        )

# ------------------------------------------------
# START THE APPLICATION
# ------------------------------------------------

demo.launch()
```
